## Part 2: Introduction to Prompt-learning

---


Prompt is the technique of making better use of the knowledge from the pre-trained model by adding additional texts to the input.

This approach significantly improves the efficiency of the use of pre-trained models, as shown in the figure below:

On the left is the traditional Model Tuning paradigm: for different tasks, the entire pre-trained language model needs to be fine-tuned, with each task having its own set of parameters.
On the right is the Prompt Tuning paradigm, where only different prompt parameters are inserted for different tasks, and each task is trained with separate prompt parameters, without training the pre-trained language model, which greatly reduces the training time and increases the usage of the model.

# Advantages of Prompt

**Level 1.** Prompt Learning makes all NLP tasks a language model problem.

Prompt Learning can normalise all tasks to the task of pre-training a language model
Avoiding the gap between pre-training and fine-tuning, almost all NLP tasks can be used directly without the need for training data.
Can achieve results beyond fine-tuning on small sample datasets.
Makes all tasks methodologically consistent

**Level 2.** Paradigm differences between Prompt Learning and Fine-tuning

Fine-tuning adapts a pre-trained language model to a downstream task.
Prompting is the task redefinition of a downstream task to make it take advantage of the capabilities of the pre-trained language model, i.e. adapting the language model


**Level 3.** The Fourth Paradigm of Modern NLP
The Prompting approach is now the fourth paradigm of NLP. The history of NLP now includes:


* Feature Engineering: i.e. using text features such as lexicality, length, etc., and training the model using machine learning methods. (No pre-trained language models)
* Architecture Engineering: based on W2V, using a deep model with a fixed embedding (with a fixed pre-trained embedding, but not directly related to the downstream task)
* Objective Engineering: Using dynamic embedding on top of bert with fine-tuning (with pre-trained language model, but with gap to downstream tasks)
* Prompt Engineering: directly using the same language model as the training model with specific prompts.(with pre-trained language model but no gap to the downstream task)

**Level 4.** Beyond the NLP perspective

Prompt can be used as an opportunity to connect multimodalities, such as the CLIP model, which connects text and images. Nowadays, sound and video can be connected, which is a vast area to be explored.

Introduction with an Example

How to construct a prompt-learning pipeline? With the modularity and flexibility of OpenPrompt, you can easily develop a prompt-learning application step by step.

## Step 1. Define a task

The first step is to determine the current NLP task, think about what’s your data looks like and what do you want from the data! That is, the essence of this step is to determine the classses and the InputExample of the task. For simplicity, we use **Sentiment Analysis** as an example.

You can also use our pre-defined Data Processors to get train/dev/test dataset for a given task.

```python
!pip install transformers==4.27.1
```

```python
%cd /content/OpenPrompt
import torch
from openprompt.data_utils import InputExample
classes = [ # There are two classes in Sentiment Analysis, one for negative and one for positive
    "negative",
    "positive"
]
dataset = [ # For simplicity, there's only two examples
    # text_a is the input text of the data, some other datasets may have multiple input sentences in one example.
    InputExample(
        guid = 0,
        text_a = "Albert Einstein was one of the greatest intellects of his time.",
    ),
    InputExample(
        guid = 1,
        text_a = "The film was badly made.",
    ),
]
```

## Step 2. Obtain a PLM

Choose a PLM to support your task. Different models have different attributes, we encourge you to use OpenPrompt to explore the potential of various PLMs. OpenPrompt is compatible with models on huggingface, the following models have been tested:

Masked Language Models (MLM): BERT, RoBERTa, ALBERT

Autoregressive Language Models (LM): GPT, GPT2

Sequence-to-Sequence Models (Seq2Seq): T5

Simply use a get_model_class to obtain your PLM.

```python
from openprompt.plms import load_plm
plm, tokenizer, model_config, WrapperClass = load_plm("bert", "bert-base-cased")
```

## Step 3. Define a Template

A Template is a modifier of the original input text, which is also one of the most important modules in prompt-learning. A more advanced tutorial to define a template is in How to Write a Template?

Here is an example, where the <text_a> will be replaced by the text_a in InputExample, and the <mask> will be used to predict a label word.

```python
from openprompt.prompts import ManualTemplate
promptTemplate = ManualTemplate(
    text = '{"placeholder":"text_a"} It was {"mask"}',
    tokenizer = tokenizer,
)
```

## Step 4. Define a Verbalizer

A Verbalizer is another important (but not necessary such as in generation) in prompt-learning,which projects the original labels (we have defined them as classes, remember?) to a set of label words. A more advanced tutorial to define a verbalizer is in How to Write a Verbalizer?

Here is an example that we

project the negative class to the word bad

project the positive class to the words good, wonderful, great.

```python
from openprompt.prompts import ManualVerbalizer
promptVerbalizer = ManualVerbalizer(
    classes = classes,
    label_words = {
        "negative": ["bad"],
        "positive": ["good", "wonderful", "great"],
    },
    tokenizer = tokenizer,
)
```

## Step 5. Construct a PromptModel

Given the task, now we have a PLM, a Template and a Verbalizer, we combine them into a PromptModel.

Note that although this example naively combine the three modules, you can actually define some complicated interactions among them.

```python
from openprompt import PromptForClassification
promptModel = PromptForClassification(
    template = promptTemplate,
    plm = plm,
    verbalizer = promptVerbalizer,
)
```

## Step 6. Define a DataLoader

A PromptDataLoader is basically a prompt version of pytorch Dataloader, which also includes a Tokenizer and a Template.

```python
from openprompt import PromptDataLoader
data_loader = PromptDataLoader(
    dataset = dataset,
    tokenizer = tokenizer,
    template = promptTemplate,
    tokenizer_wrapper_class=WrapperClass,
)
```

## Step 7. Train and inference
Done! We can conduct training and inference the same as other processes in Pytorch.

This is a quick start of OpenPrompt, please refer to the APIs for more details.

```python
import torch

# making zero-shot inference using pretrained MLM with prompt
promptModel.eval()
with torch.no_grad(): # inference
    for batch in data_loader: #batch
        logits = promptModel(batch)
        preds = torch.argmax(logits, dim = -1)
        print(classes[preds])
# predictions would be 1, 0 for classes 'positive', 'negative'
```
